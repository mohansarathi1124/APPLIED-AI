```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import tensorflow as tf 
from sklearn.model_selection import train_test_split, StratifiedKFold 
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import (confusion_matrix, classification_report, roc_auc_score, 
                             precision_recall_curve, auc, roc_curve) 
from sklearn.utils.class_weight import compute_class_weight 

# SETTING UNIQUE SEED FOR YOUR RESULTS
np.random.seed(101) 
tf.random.set_seed(101) 

# --- DATA PREPARATION ---
df = pd.read_csv("creditcard.csv") 
X = df.drop("Class", axis=1) 
y = df["Class"] 

# 1. VISUALIZING CLASS IMBALANCE (The missing graph)
plt.figure(figsize=(8, 5))
counts = df['Class'].value_counts()
counts.plot(kind='bar', color=['blue', 'red'], rot=0)
plt.title('Transaction Class Distribution')
plt.xticks(range(2), ['Normal (0)', 'Fraud (1)'])
plt.xlabel('Class')
plt.ylabel('Number of Transactions')
for i, v in enumerate(counts):
    plt.text(i - 0.1, v + 3000, str(v), color='black', fontweight='bold')
plt.show()

# Data Scaling
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X) 

# Handle Class Imbalance Weights
classes = np.unique(y) 
class_weights = compute_class_weight(class_weight="balanced", classes=classes, y=y) 
class_weight_dict = dict(zip(classes, class_weights)) 

# --- DNN ARCHITECTURE ---
def build_dnn(input_dim): 
    model = tf.keras.Sequential([ 
        tf.keras.Input(shape=(input_dim,)), 
        tf.keras.layers.Dense(128, activation="relu", 
                             kernel_regularizer=tf.keras.regularizers.l2(1e-4)), 
        tf.keras.layers.BatchNormalization(), 
        tf.keras.layers.Dropout(0.3), 
        tf.keras.layers.Dense(64, activation="relu", 
                             kernel_regularizer=tf.keras.regularizers.l2(1e-4)), 
        tf.keras.layers.BatchNormalization(), 
        tf.keras.layers.Dropout(0.3), 
        tf.keras.layers.Dense(32, activation="relu", 
                             kernel_regularizer=tf.keras.regularizers.l2(1e-4)), 
        tf.keras.layers.Dense(1, activation="sigmoid") 
    ]) 
    
    # UNIQUE LEARNING RATE (0.0005)
    model.compile( 
        optimizer=tf.keras.optimizers.Adam(learning_rate=5e-4), 
        loss="binary_crossentropy", 
        metrics=["accuracy"] 
    ) 
    return model 

# Evaluation & Plotting Functions
def evaluate_model(model, X_test, y_test): 
    y_prob = model.predict(X_test).ravel() 
    y_pred = (y_prob > 0.5).astype(int) 
    roc_auc = roc_auc_score(y_test, y_prob) 
    precision, recall, _ = precision_recall_curve(y_test, y_prob) 
    pr_auc = auc(recall, precision) 
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred)) 
    print("Classification Report:\n", classification_report(y_test, y_pred, digits=4)) 
    return roc_auc, pr_auc, y_prob 

def plot_curves(y_test, y_prob, title): 
    fpr, tpr, _ = roc_curve(y_test, y_prob) 
    precision, recall, _ = precision_recall_curve(y_test, y_prob) 
    plt.figure(figsize=(12, 5)) 
    plt.subplot(1, 2, 1) 
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(y_test, y_prob):.4f}') 
    plt.plot([0, 1], [0, 1], linestyle="--") 
    plt.xlabel("False Positive Rate") 
    plt.ylabel("True Positive Rate") 
    plt.title(f"ROC Curve – {title}") 
    plt.legend()
    plt.subplot(1, 2, 2) 
    plt.plot(recall, precision) 
    plt.xlabel("Recall") 
    plt.ylabel("Precision") 
    plt.title(f"PR Curve – {title}") 
    plt.tight_layout() 
    plt.show() 

# --- EXPERIMENTS ---

# Experiment 1: 80–20 Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=101) 
model1 = build_dnn(X_train.shape[1]) 
model1.fit(X_train, y_train, epochs=30, batch_size=512, class_weight=class_weight_dict, verbose=1) 
roc1, pr1, y_prob1 = evaluate_model(model1, X_test, y_test) 
plot_curves(y_test, y_prob1, "Experiment 1 (80–20)") 

# Experiment 2: 70–30 Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=101) 
model2 = build_dnn(X_train.shape[1]) 
model2.fit(X_train, y_train, epochs=30, batch_size=512, class_weight=class_weight_dict, verbose=1) 
roc2, pr2, y_prob2 = evaluate_model(model2, X_test, y_test) 
plot_curves(y_test, y_prob2, "Experiment 2 (70–30)") 

# Experiment 3: 60-20-20 Split
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=101)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=101)
early_stop = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
model3 = build_dnn(X_train.shape[1])
model3.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=512, class_weight=class_weight_dict, callbacks=[early_stop], verbose=1)
roc3, pr3, y_prob3 = evaluate_model(model3, X_test, y_test)
plot_curves(y_test, y_prob3, "Experiment 3 (60-20-20)")

# Experiment 4: 70-15-15 Split
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.15, stratify=y, random_state=101)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=101)
model4 = build_dnn(X_train.shape[1])
model4.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=512, class_weight=class_weight_dict, callbacks=[early_stop], verbose=1)
roc4, pr4, y_prob4 = evaluate_model(model4, X_test, y_test)
plot_curves(y_test, y_prob4, "Experiment 4 (70-15-15)")
```
