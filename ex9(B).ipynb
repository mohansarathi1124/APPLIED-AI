```python
# Neural Machine Translation (English → French) using Transformer

# Install packages (run once)
# !pip install transformers datasets sentencepiece sacrebleu torch

from datasets import load_dataset
from transformers import MarianMTModel, MarianTokenizer
import sacrebleu

# 1. Load dataset
dataset = load_dataset("opus100", "en-fr")

# Use 150 sentences for evaluation
test_data = dataset["test"].select(range(150))

# 2. Load pretrained translation model
model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

predictions = []
references = []

# 3. Translate sentences
for item in test_data:
    
    english_sentence = item["translation"]["en"]
    french_reference = item["translation"]["fr"]

    inputs = tokenizer(english_sentence, return_tensors="pt", truncation=True)

    translated = model.generate(**inputs)

    predicted_fr = tokenizer.decode(translated[0], skip_special_tokens=True)

    predictions.append(predicted_fr)
    references.append([french_reference])

# 4. Compute BLEU Score
bleu = sacrebleu.corpus_bleu(predictions, references)

print("\nBLEU Score:", bleu.score)

# 5. Show sample translations
print("\nSample Translations:\n")

for i in range(5):
    print("English :", test_data[i]["translation"]["en"])
    print("Reference French :", test_data[i]["translation"]["fr"])
    print("Predicted French :", predictions[i])
    print("-"*50)
```

```python
# !pip install transformers datasets sentencepiece sacrebleu torch

from datasets import load_dataset
from transformers import MarianMTModel, MarianTokenizer
import sacrebleu

# 1. Load dataset
dataset = load_dataset("opus100", "en-fr")

# Use 200 samples
test_data = dataset["test"].select(range(200))

# 2. Load pretrained model
model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

predictions = []
references = []

# 3. Translate
for item in test_data:

    en = item["translation"]["en"]
    fr = item["translation"]["fr"]

    inputs = tokenizer(en, return_tensors="pt", truncation=True)

    output = model.generate(**inputs)

    pred = tokenizer.decode(output[0], skip_special_tokens=True)

    predictions.append(pred)
    references.append(fr)

# 4. Format references correctly for sacrebleu
bleu = sacrebleu.corpus_bleu(predictions, [references])

print("BLEU Score:", bleu.score)

# 5. Show examples
for i in range(5):
    print("\nEnglish:", test_data[i]["translation"]["en"])
    print("Reference:", references[i])
    print("Prediction:", predictions[i])
```
